```python
# Добавим новые фичи: Новый год, ковид и рамадан
def covid(ds):
    return (ds >= pd.to_datetime('2020-03-18') and ds <= pd.to_datetime('2020-05-11'))
    
def new_year(ds):
    return ds in [pd.to_datetime('2020-01-01'), pd.to_datetime('2020-01-02')] \
        or ds in [pd.to_datetime('2021-01-01'), pd.to_datetime('2021-01-02')] \
        or ds in [pd.to_datetime('2022-01-01'), pd.to_datetime('2022-01-02')] \
        or ds in [pd.to_datetime('2023-01-01'), pd.to_datetime('2023-01-02')] \

def ramadan(ds):
    return (ds >= pd.to_datetime("2020-04-24") and ds <= pd.to_datetime("2020-05-24")
        or ds >= pd.to_datetime("2021-04-13") and ds <= pd.to_datetime("2021-05-13")
        or ds >= pd.to_datetime("2022-04-03") and ds <= pd.to_datetime("2022-05-03")
        or ds >= pd.to_datetime("2023-03-23") and ds <= pd.to_datetime("2023-04-24"))
    
data['covid'] = data['ds'].apply(covid).astype('int')
data['new_year'] = data['ds'].apply(new_year).astype('int')
data['ramadan'] = data['ds'].apply(ramadan).astype('int')
```

```python
# Тестим с новыми фичами
train_data = data[:731]
train_X = train_data[['month', 'day_of_week', 'days_passed', 'covid', 'new_year', 'ramadan']] #2020, 2021
train_y = train_data['y']

test_data = data[731:]
test_X = test_data[['month', 'day_of_week', 'days_passed', 'covid', 'new_year', 'ramadan']] #2022
test_y = test_data['y']

model = MyLineReg(metric='mape')
model.fit(train_X, train_y)
plot(model)
```

```python
# Провалидируем модель на 2022 году
prediction = pd.DataFrame(model.predict(test_X), columns=['yhat'])
actual = pd.DataFrame(test_y, columns=['y'])
prediction.index = actual.index
mean_absolute_percentage_error(prediction['yhat'], actual['y'])
```

```python
actual.y.plot(figsize=(10,5))
prediction.yhat.plot(figsize=(10,5))
```

```python
#Модель сильно ошибается в последний день 2022 года. Заменим данные на среднее в месяце.
data.loc['2022-12-31', 'y'] = int(np.mean(data.query('ds.dt.year == 2020 & ds.dt.month == 12').y))
```

```python
# Переберем параметры
train_data = data
train_X = train_data[['month', 'day_of_week', 'days_passed', 'covid', 'new_year', 'ramadan']]
train_y = train_data['y']

def train(params):
    model = MyLineReg(metric='mape', **params)
    model.fit(train_X, train_y)

    prediction = pd.DataFrame(model.predict(train_X), columns=['yhat'])
    actual = pd.DataFrame(train_y, columns=['y'])  
    
    return model.get_best_score()
```

```python
param_grid = {
    'n_iter': np.arange(2000, 10001, 500),
    'learning_rate': np.arange(0.01, 0.06, 0.01)
    # 'reg': ['elasticnet', None, 'l1'],
    # 'l1_coef': np.arange(0, 1.1, 0.1)
}
all_params = [dict(zip(param_grid.keys(), v)) for v in itertools.product(*param_grid.values())]

mapes = []

start_time = time()

for params in tqdm(all_params):
    mapes.append(train(params))
best_params = all_params[np.argmin(mapes)]
print('Total params: ', len(all_params))
print('Time used: ', time() - start_time)
print(best_params)
'MAPE', min(mapes)
```
